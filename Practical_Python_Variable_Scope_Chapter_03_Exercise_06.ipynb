### The Confusion: When Is It Shadowing vs Reassignment?

The confusion arises when you try to modify a variable from an outer scope. Python's rule is simple but can be surprising:

**If you assign to a variable name anywhere in a function, that name is considered local for the entire function.**

This means you can't mix reading and writing to the same name:

```python
x = 10  # Global

def problematic():
    print(x)  # Try to read global x
    x = 20    # Try to create local x
    # This causes UnboundLocalError!

problematic()  # UnboundLocalError: local variable 'x' referenced before assignment
```

Here's what Python does:

1. Python analyzes the entire function before executing it
2. It sees `x = 20` and decides: "`x` must be a local variable for this entire function"
3. When `print(x)` executes, Python looks for `x` in the local namespace first
4. The local `x` doesn't exist yet (because `x = 20` hasn't executed)
5. Python raises `UnboundLocalError`

This is shadowing in actionâ€”Python is trying to create a local `x` that shadows the global `x`, but you're trying to read it before it's created.
