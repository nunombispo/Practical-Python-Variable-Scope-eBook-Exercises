# Chapter 3: Local Scope (Functions)

## Exercise 05

**Reassignment** is when you modify a variable that exists in the current scope. For local variables, this is straightforward:

```python
def my_function():
    x = 10      # Create local variable
    x = 20      # Reassign to the same local variable
    x = 30      # Reassign again
    return x

print(my_function())  # 30
```

All three assignments modify the same local variable `x`. This is reassignment within the local scope.
