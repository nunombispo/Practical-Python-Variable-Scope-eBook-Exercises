# Chapter 2: Namespaces & the LEGB Rule

## Exercise 08

`locals()` returns a dictionary representing the current local namespace. In a function, this shows all local variables. At module level, it's the same as `globals()`.

```python
def my_function():
    x = 10
    y = 20
    print("Local namespace:", locals())

my_function()
# Output: Local namespace: {'x': 10, 'y': 20}
```

`globals()` returns a dictionary representing the current global namespace (module-level namespace).

```python
x = "global_x"
y = "global_y"

def my_function():
    local_var = "local"
    print("x in globals:", 'x' in globals())
    print("local_var in globals:", 'local_var' in globals())

my_function()
# Output:
# x in globals: True
# local_var in globals: False
```

Unlike `locals()`, modifying `globals()` **does** modify the actual global namespace:

```python
x = 10

def modify_global():
    globals()['x'] = 20  # This actually modifies the global x

modify_global()
print(x)  # 20 (was modified)
```

However, using the `global` keyword is the preferred and clearer way to modify globals.
