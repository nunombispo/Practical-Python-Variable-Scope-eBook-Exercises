# Chapter 3: Local Scope (Functions)

## Exercise 04

**Shadowing** occurs when you assign to a variable name that already exists in an outer scope. Instead of modifying the outer variable, Python creates a new local variable with the same name, which "shadows" (hides) the outer one.

```python
x = "global"  # Global variable

def my_function():
    x = "local"  # This creates a NEW local variable, shadowing the global
    print(x)     # Prints "local"

my_function()
print(x)  # Prints "global" - the global variable was NOT modified
```

Here's what happens step by step:

1. The global namespace has `x = "global"`
2. When `my_function()` is called, Python sees `x = "local"` and decides `x` must be local
3. Python creates a new local variable `x = "local"` in the function's local namespace
4. This local `x` shadows (hides) the global `x` within the function
5. When the function returns, the local `x` is destroyed
6. The global `x` remains unchanged

**Important**: The assignment `x = "local"` does **not** modify the global `x`. It creates a completely separate local variable.
