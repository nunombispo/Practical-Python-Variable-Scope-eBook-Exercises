# Chapter 2: Namespaces & the LEGB Rule

## Exercise 05

The **built-in scope** contains Python's built-in names like `print`, `len`, `str`, `range`, `Exception`, etc. This is the fallback scope, if Python doesn't find a name in Local, Enclosing, or Global, it checks the built-in scope.

```python
def my_function():
    # No local, enclosing, or global 'len'
    # So Python uses built-in len()
    return len([1, 2, 3])

print(my_function())  # 3
```

**Key characteristics:**

- Always available
- Contains all of Python's built-in functions and exceptions
- Can be shadowed (overridden) by local, enclosing, or global names
- Last in the search order (LEGB)

You can shadow built-ins, but it's usually a bad idea:

```python
def my_function():
    str = "I shadowed str!"  # Bad practice: shadows built-in str
    # Now you can't use str() as a function in this scope

my_function()
```
