# Chapter 2: Namespaces & the LEGB Rule

## Exercise 02

The **local scope** is the innermost scope, the current function or method you're executing. Variables assigned directly in a function (without `global` or `nonlocal`) are local to that function.

```python
def my_function():
    x = 10  # Local variable
    print(x)  # Uses local x

my_function()
```

**Key characteristics:**

- Created when the function is called
- Destroyed when the function returns
- Each function call gets its own local namespace
- Local variables are private to the function
