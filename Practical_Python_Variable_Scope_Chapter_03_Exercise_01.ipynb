# Chapter 3: Local Scope (Functions)

## Exercise 01

A **local variable** is any variable that's **assigned** inside a function (without using `global` or `nonlocal`). When you write a function, any variable you assign to becomes local to that function by default.

```python
def calculate_total(price, tax_rate):
    subtotal = price  # Local variable
    tax = subtotal * tax_rate  # Local variable
    total = subtotal + tax  # Local variable
    return total

result = calculate_total(100, 0.08)
# After the function returns, subtotal, tax, and total no longer exist
```

**Key characteristics of local variables:**

1. **Created on function call**: Local variables are created when the function is called, not when it's defined
2. **Private to the function**: Each function call gets its own independent set of local variables
3. **Destroyed on return**: When the function returns, all local variables are destroyed
4. **No interference**: Multiple calls to the same function don't interfere with each other's local variables

Let's see this independence in action:

```python
def counter():
    count = 0  # Local variable
    count += 1
    return count

print(counter())  # 1
print(counter())  # 1 (not 2!)
print(counter())  # 1 (not 3!)
```

Each call to `counter()` creates a fresh `count` variable initialized to 0. The previous call's `count` is goneâ€”it was destroyed when that function call returned.
