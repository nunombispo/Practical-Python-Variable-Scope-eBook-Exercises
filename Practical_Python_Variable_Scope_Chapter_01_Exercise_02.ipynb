## Problem 1: Silent Failures

Many scope bugs don't crash your program. Instead, they produce incorrect results silently:

```python
data = []

def add_item(item):
    data.append(item)

add_item(1)
add_item(2)
print(data)  # [1, 2] - works fine!

# But what if someone tries to reset the data?
def reset_data():
    data = []  # Oops! Creates a local variable, not modifying global

reset_data()
print(data)  # Still [1, 2] - the reset didn't work!
```

The `reset_data()` function doesn't crash, but it doesn't work either. The assignment `data = []` creates a new local variable instead of modifying the global one. The local variable is destroyed when the function ends, leaving the global `data` unchanged.

**This is dangerous** because your code runs without errors, but produces wrong results. You might not notice the bug until much later, making it very hard to debug.
