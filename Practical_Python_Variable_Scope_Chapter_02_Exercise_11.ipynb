### Misconception 2: "I need `global` to read a global variable"

You only need `global` (or `nonlocal`) to **modify** a non-local variable. Reading works automatically through LEGB lookup.

```python
x = 10

def read_global():
    print(x)  # Works fine without 'global' keyword

def modify_global():
    global x  # Need this to modify
    x = 20

read_global()  # 10
modify_global()
read_global()  # 20
```
