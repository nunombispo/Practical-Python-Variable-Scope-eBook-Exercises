## Exercise 09

While you won't use these functions in everyday code, they're valuable for:

1. **Debugging**: Understanding what variables are available in a scope
2. **Learning**: Seeing how Python organizes namespaces

```python
def debug_function(a, b):
    """Helper function to see all variables in scope"""
    print("Function parameters and locals:")
    for name, value in locals().items():
        print(f"  {name} = {value}")

debug_function(1, 2)
# Output:
# Function parameters and locals:
#   a = 1
#   b = 2
```
