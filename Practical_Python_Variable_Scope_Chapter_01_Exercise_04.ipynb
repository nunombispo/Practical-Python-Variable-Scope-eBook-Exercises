## Problem 3: Context-Dependent Behavior

A function that works in one situation might break in another, depending on what variables exist:

```python
# This function works fine when tested alone
def log_event(event_name):
    events.append(event_name)  # Modifies global list
    return len(events)

# Test it:
events = []
log_event("click")  # Works: returns 1
log_event("hover")  # Works: returns 2
print(events)  # ['click', 'hover']

# But if someone adds this later:
def log_event(event_name):
    if len(events) > 100:  # Check global first
        events = []        # Reset - but this makes 'events' local!
    events.append(event_name)  # Now this fails!
    return len(events)

# When used with a global events list, it crashes:
events = ['old_event']
log_event("new")  # UnboundLocalError!
```

The function worked fine initially, but when someone tries to improve it by checking the global variable first, it breaks. This happens because Python sees the assignment `events = []` and marks `events` as local for the entire function, so the check `len(events)` tries to read a local variable that doesn't exist yet.
