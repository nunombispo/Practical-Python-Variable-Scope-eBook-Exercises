# Chapter 2: Namespaces & the LEGB Rule

## Exercise 07

An important distinction: Python treats **reading** and **writing** variables differently when resolving scope.

**Reading** (just using a variable):

- Python searches LEGB automatically
- No keywords needed
- Stops at first match

```python
x = "global"

def my_function():
    print(x)  # Reads global x - works fine

my_function()  # Prints "global"
```

**Writing** (assigning to a variable):

- Python assumes the variable is local by default
- If you want to modify a non-local variable, you need `global` or `nonlocal`
- This is why you get `UnboundLocalError` when trying to modify before assigning

```python
x = "global"

def my_function():
    x = "local"  # Writing creates a local variable
    # This shadows the global x, but doesn't modify it

my_function()
print(x)  # Still "global" - the global wasn't modified
```
