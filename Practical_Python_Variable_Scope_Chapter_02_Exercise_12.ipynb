### Misconception 3: "All variables in a function are local"

Only variables that are **assigned to** in a function are considered local. Variables that are only read can come from any scope.

```python
x = 10
y = 20

def my_function():
    print(x)    # Reads global x - not local
    y = 30      # Assigns to y - this creates a local y
    print(y)    # Prints local y (30)

my_function()
print(y)  # Global y is still 20
```
