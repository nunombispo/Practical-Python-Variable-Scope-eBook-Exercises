## Problem 2: Immediate Crashes

Other scope bugs cause immediate crashes, which at least makes them easier to spot:

```python
counter = 0

def increment():
    counter = counter + 1  # UnboundLocalError!
    return counter

increment()  # Crashes immediately
```

This crashes because Python sees `counter = counter + 1` and decides `counter` must be local. But then when it tries to evaluate `counter + 1`, it needs the value of `counter` first, which doesn't exist yet in the local scope.

At least with immediate crashes, you know something is wrong right away!
