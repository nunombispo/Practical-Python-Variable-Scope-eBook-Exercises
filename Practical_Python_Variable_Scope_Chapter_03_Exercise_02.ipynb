# Chapter 3: Local Scope (Functions)

## Exercise 02

Function parameters are also local variables. They're just local variables that get their initial values from the function call:

```python
def greet(name):
    # 'name' is a local variable, initialized from the function argument
    message = f"Hello, {name}!"  # Another local variable
    return message

greet("Alice")  # name = "Alice" (local)
greet("Bob")    # name = "Bob" (a different local variable in a different call)
```
