## Exercise 04

The **global scope** is the module-level namespace. Variables defined at the top level of a module (not inside any function or class) are in the global scope.

```python
x = "global"  # Global variable

def my_function():
    print(x)  # Accesses global x

my_function()  # Prints "global"
```

**Key characteristics:**

- Created when the module is imported
- Exists for the lifetime of the program (or until module reload)
- Accessible from anywhere in the module
- Modifiable from anywhere (with `global` keyword)
