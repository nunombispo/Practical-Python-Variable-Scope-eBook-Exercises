# Chapter 2: Namespaces & the LEGB Rule

## Exercise 06

Let's see the LEGB rule in action with one comprehensive example that shows all four levels:

```python
# Built-in: print, len, etc. (B)
x = "global"  # Global (G)

def outer():
    x = "enclosing"  # Enclosing (E)

    def inner():
        x = "local"  # Local (L)
        print(x)  # Which x? Local (stops at L)

    inner()

outer()  # Prints "local"
```

Now let's see what happens when we remove variables, one level at a time:

```python
x = "global"  # Global

def outer():
    x = "enclosing"  # Enclosing

    def inner():
        # No local x, so Python searches enclosing
        print(x)  # Which x? Enclosing (stops at E)

    inner()

outer()  # Prints "enclosing"
```

```python
x = "global"  # Global

def outer():
    def inner():
        # No local or enclosing x, so Python searches global
        print(x)  # Which x? Global (stops at G)

    inner()

outer()  # Prints "global"
```

```python
def my_function():
    # No local, enclosing, or global 'len'
    # So Python uses built-in len (stops at B)
    return len([1, 2, 3])

print(my_function())  # 3 (uses built-in len)
```
