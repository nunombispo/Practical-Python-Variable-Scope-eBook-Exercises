### Misconception 1: "Variables are looked up at runtime dynamically"

While Python is a dynamic language, variable lookup follows the LEGB rule **lexically** (based on the code structure), not dynamically (based on the call stack). The scope where a variable is found is determined by where the code is written, not where it's called from.

```python
x = "global"

def outer():
    x = "enclosing"

    def inner():
        print(x)  # Always uses enclosing x, regardless of call site

    return inner

func = outer()

# Even though we're calling from global scope, inner() still uses enclosing x
func()  # Prints "enclosing", not "global"
```
