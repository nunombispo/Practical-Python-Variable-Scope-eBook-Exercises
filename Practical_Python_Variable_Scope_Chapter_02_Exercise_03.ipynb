## Exercise 03

The **enclosing scope** (also called nonlocal scope) refers to the scope of any enclosing functions. This comes into play with nested functions.

```python
def outer():
    x = "enclosing"  # Enclosing scope variable

    def inner():
        print(x)  # Accesses enclosing scope

    inner()

outer()  # Prints "enclosing"
```

**Key characteristics:**

- Only exists when you have nested functions
- Inner functions can read enclosing variables
- Inner functions cannot modify enclosing variables without `nonlocal` (we'll cover this in Chapter 5)
- The search goes outward through all enclosing scopes
