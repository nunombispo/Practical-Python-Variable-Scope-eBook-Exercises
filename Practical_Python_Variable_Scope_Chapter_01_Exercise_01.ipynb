# Chapter 1: Why Variable Scope Matters

## Exercise 01

When you write `x = 10` in Python, you're not just creating a value. You're creating a name binding in a specific namespace, and that namespace determines where `x` can be seen and modified.

Consider this simple example:

```python
x = 5

def my_function():
    print(x)

my_function()  # Prints: 5
```

This works because `x` is in the global scope, and functions can read variables from the global scope. But what if we try to modify `x`?

```python
x = 5

def my_function():
    print(x)
    x = 10  # Try to change x

my_function()  # UnboundLocalError!
```

Now you'll get an `UnboundLocalError`, even though `x` clearly exists globally. This is scope in action.

**Why this happens:** Python analyzes the entire function before executing it. When it sees `x = 10` anywhere in the function, it decides that `x` must be a local variable for that entire function. So when `print(x)` executes, Python looks for `x` in the local namespace first, doesn't find it yet (because `x = 10` hasn't executed), and raises an error.
