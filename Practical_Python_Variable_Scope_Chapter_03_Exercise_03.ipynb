# Chapter 3: Local Scope (Functions)

## Exercise 03

Each function call creates its own local namespace. This means variables with the same name in different calls are completely separate:

```python
def process_data():
    data = []  # Local variable
    data.append(1)
    return data

result1 = process_data()  # Creates local 'data', returns [1]
result2 = process_data()  # Creates a NEW local 'data', returns [1]
# The two 'data' variables are completely independent
```
