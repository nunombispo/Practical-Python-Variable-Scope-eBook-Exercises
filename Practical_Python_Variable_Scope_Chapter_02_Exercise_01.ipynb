# Chapter 2: Namespaces & the LEGB Rule

## Exercise 01

Python has multiple namespaces active at any given time:

```python
# Global namespace (module-level)
x = 10

def my_function():
    # Local namespace (function-level)
    y = 20
    return x + y

my_function()
```

Here, `x` lives in the global namespace, and `y` lives in the local namespace of `my_function`. When Python executes `x + y`, it needs to find both variables. How does it know where to look?

The answer is **scope resolution**, the process of determining which namespace contains a given name. Python uses the **LEGB rule** to resolve this systematically.
